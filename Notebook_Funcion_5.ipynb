Se importa la librería necesaria

```python
import pandas as pd
```

Se utiliza el dataframe 'user_review_sentiment', así que se carga.

```python
df1 = pd.read_csv('user_review_sentiment.csv')
df1 = df1[['user_id', 'item_id', 'sentiment']]
df1
```

Renombra la columna 'item_id'

```python
df1 = df1.rename(columns={'item_id': 'id'})
```

Se revisa el dataFrame

```python
df1
```

leemos el archivo csv que pertenece a 'steam games'

```python
df2 = pd.read_csv('steam_games.csv')
df2 = df2[['id', 'release_date']]
df2
```

Eliminamos datos NaN del dataframe

```python
df2 = df2.dropna()
df2
```

Conviertimos la columna 'release_date' en objetos de fecha, extrae el año de estas fechas y crea una nueva columna llamada 'release_year' que contiene el año de lanzamiento de cada entrada en el DataFrame df2.

```python
# Convierte la columna 'release_date' a objetos de fecha con el formato 'YYYY-MM-DD'
df2['release_date'] = pd.to_datetime(df2['release_date'], format='%Y-%m-%d', errors='coerce')

# Extrae el año y crea una nueva columna 'release_year'
df2['release_year'] = df2['release_date'].dt.year

df2['release_year'] 

```

Elimino la columna 'release_date' del DataFrame

```python

df2 = df2.drop('release_date', axis=1)
```

aseguramos que no haya valores nulos en la columna 'release_year' y luego la convierte a tipo entero para su posterior análisis o procesamiento.

```python
#convierto el año en un número entero
df2['release_year'].fillna(0, inplace=True)

# Convert the 'release_year' column to integer type
df2['release_year'] = df2['release_year'].astype(int)
```

```python
df2
```

Se hace merge entre los dos dataFrames

```python
df_merged = df1.merge(df2, on='id', how='left')
```

```python
df_merged
```

Se asegura que no hayan valores NaN

```python
df_merged = df_merged.dropna(subset=['release_year'])
```

Se convierte a entero todos los registros de la columna 'release_year'

```python
#convierto el año en un número entero
df2['release_year'] = df2['release_year'].astype(int)
```

```python
df_merged = df_merged [['sentiment', 'release_year']]
df_merged
```

visualizamos la distribución de los valores de sentimiento para cada año en el conjunto de datos.

```python
# Crear una tabla cruzada (crosstab) para contar los valores de sentiment por año
crosstab_result = pd.crosstab(df_merged['release_year'], df_merged['sentiment'])

# Renombrar las columnas para mayor claridad
crosstab_result.columns = ['sentiment_0', 'sentiment_1', 'sentiment_2']

# Mostrar el resultado
crosstab_result
```

Se guarda la tabla cruzada

```python
crosstab_result.to_csv('release_year_sentiment_0-1-2.csv')
```

Se crea un dataframe apartir del csv guardado

```python
df4 = pd.read_csv('release_year_sentiment_0-1-2.csv')
```

 realizamos algunas manipulaciones de datos en el DataFrame df4, lo guardamos como un archivo CSV y luego mostramos el DataFrame resultante. Las columnas 'sentiment_0', 'sentiment_1' y 'sentiment_2' se renombran como 'Negativo', 'Neutro' y 'Positiva', respectivamente, y la columna 'release_year' se convierte en un entero. Finalmente, el DataFrame modificado se guarda como un archivo CSV.

```python
df4['release_year'] = df4['release_year'].astype(int)
df4.rename(columns={'sentiment_0': 'Negativo'}, inplace=True)
df4.rename(columns={'sentiment_1': 'Neutro'}, inplace=True)
df4.rename(columns={'sentiment_2': 'Positiva'}, inplace=True)
df4.to_csv('Funcion_5_jup.csv')
df4
```

```python

sentiment_analysis = pd.read_csv('Funcion_5_jup.csv')
sentiment_analysis  = sentiment_analysis.drop('Unnamed: 0', axis=1)
def Sentiment_Analysis(release_year):
    # Filtrar el DataFrame para obtener las filas que coinciden con el release_year dado
    filas_coincidentes = sentiment_analysis[sentiment_analysis['release_year'] == release_year]
    
    # Convertir las filas coincidentes a una lista de diccionarios
    filas_lista = filas_coincidentes.to_dict(orient='records')
    
    return filas_lista
```
